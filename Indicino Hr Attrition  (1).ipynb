## Hr Attrition Using Machine Learning- Classification

```python
# Importing the necessary libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Hr dataset

df= pd.read_excel(r"C:\Users\Daniel\Downloads\Indicino project.xlsx")
```

```python
#Displaying the first 10 rows of the dataset

df[0:10]
```

### Data Preprocessing
#### Check for missing values and duplicates

```python
# Checking for missing vaues

df.isna().sum()
```

```python
# Checking for duplicates

df.duplicated().sum()
```

```python
df.info()
```

#### Convert categorical columns to Numeric
##### Using Label Encoding 

```python
# identifying all the caegorical columns

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:", categorical_cols)
```

```python
# Encoding the categorical columns

encoder= LabelEncoder()

df['Attrition'] = encoder.fit_transform(df['Attrition'])  #convert yes/no to 1,0

#Converting other categorical columns
categorical_colls= ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender',
       'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
for cols in categorical_colls:
    df[cols]= encoder.fit_transform(df[cols])
```

#### Features selection

```python
# Features
X= df.drop(columns=['Attrition', 'StandardHours', 'EmployeeNumber', 'Over18', 'EmployeeCount'], axis=1)
X.info()
```

```python
# Target
y= df['Attrition']
```

##### Data Imbalance Check & Oversampling

```python
# Checking for data imbalance in the Target variable

y.value_counts().plot(kind='bar')
plt.show()
```

```python
# Over sampling using SMOTE(synthetic minority oversampling techniqe)
# Oversampling the minority class will help increase its representation in the dataset, 
# making it easier for the model to learn the patterns and characteristics of that class.
```

```python
# importing SMOTE for over_sampling

from imblearn.over_sampling import SMOTE
```

```python
# Now resample

smote= SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled= smote.fit_resample(X,y)
```

```python
# visualize if the target data is now balanced

y_resampled.value_counts().plot(kind='bar')
plt.show()
```

```python
# Standardize numerical features

scaler= StandardScaler()
X_scaled= scaler.fit_transform(X_resampled)
```

#### Split data into Train and Test set

```python
X_train, X_test, y_train, y_test= train_test_split(X_scaled, y_resampled, test_size= 0.2, random_state= 42)
```

#### Train the model using RandomForestClassifier

```python
model= RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

#### Evaluate Model performance Using Metrics
##### Accuracy

```python
# Predicting x_train
train_pred= model.predict(X_train)
```

```python
# evaluate the model using Accuracy_score

train_accuracy= accuracy_score(y_train, train_pred)

print(f"Model Accuracy: {train_accuracy:.2%}") #Displaying output in %
```

```python
# Predicting x_test
y_pred= model.predict(X_test)
```

```python
# evaluate the model using Accuracy_score

accuracy= accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.2%}") #Displaying output in %
```

#### Classification Report 

```python
classification= classification_report(y_test, y_pred)

print("Classification Report:\n", classification)
```

##### Confusion Matrix

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
